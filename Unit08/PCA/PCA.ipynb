```python
import numpy as np
import sklearn
from skimage import io
from skimage import color
from sklearn.decomposition import PCA

#none - origin
#n - normal
#re - reshape 1*n
#0 - origin-mean

train_img = io.ImageCollection('./p2_data/*_[1-9].png')
test_img = io.ImageCollection('./p2_data/*_10.png')
#print("Train",train_img)
#print("Test",test_img )
N = len(train_img)
img = io.imread('./p2_data/2_1.png')
w=len(img[0])
h=len(img)
#print("寬：",len(img[0]),"高：",len(img))
data = np.zeros([N,w*h])
data = np.zeros([N,56*46])
eigen_face = np.zeros([N,56,46])
```

```python
###### mean face ######
mean_img = np.mean(train_img, axis=0, dtype=float)
print(mean_img)

'''
mean_img_n = (mean_img-mean_img.min())/(mean_img.max()-mean_img.min())
io.imsave('./pca/_mean.png', mean_img_n)
'''

mean_img_re = mean_img.reshape(1,56*46)
img_re = img.reshape(1,56*46)

for i in range(len(train_img)):
    data[i] = train_img[i].reshape(1,56*46) - mean_img_re
```

```python
###### eigen face ######
pca = PCA(n_components=N)
pca.fit(data)

eigen_re = pca.components_
eigen = eigen_re.reshape(N,56,46)

for n in range(len(train_img)):
    eigen_face[n] = (eigen[n]-eigen[n].min())/(eigen[n].max()-eigen[n].min())
    io.imsave('./pca/'+ np.str(n+1) +'.png', eigen_face[n])
```

```python
###### reconstruct images ######
img0 = img_re - mean_img_re
ipca = PCA(n_components=345).fit(data)
comp = ipca.transform(img0)
new_re = ipca.inverse_transform(comp) + mean_img_re
new = np.reshape(new_re, (56,46))
new_n = (new-new.min())/(new.max()-new.min())
io.imsave('./345.png',new_n)

###### MSE ######
mse = ((new_re - img_re)**2).mean()
print("mse = ",mse)
```

```python
###### KNN validation ######
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import KFold

label = np.zeros(360)
for i in range(40):
    for j in range(9):
        label[i*9+j] = i+1
#print(label)

def KNNclassify(k, n):
    
    X = np.matmul(data, eigen_re[:n].T)
    
    kf = KFold(n_splits=3, shuffle=True)
    accuracy = 0
    
    for train_index, val_index in kf.split(X):
        X_train, X_val = X[train_index], X[val_index]
        y_train, y_val = label[train_index], label[val_index]
    
        knn = KNeighborsClassifier(n_neighbors=k)
        knn.fit(X_train, y_train)
        val_accuracy = knn.score(X_val, y_val)
    
        predict = knn.predict(X_val)
        accuracy += val_accuracy
    
    return accuracy/3

for k in [1,3,5]:
    for n in [3,50,170]:
        print("k=",k, ", n=",n, ", accuracy=", KNNclassify(k,n))
```

```python
###### KNN test ######
from sklearn.neighbors import KNeighborsClassifier

train_label = np.zeros(360)
test_label = np.zeros(40)

for i in range(40):
    for j in range(9):
        train_label[i*9+j] = i+1
        
for i in range(40):
    for j in range(1):
        test_label[i*1+j] = i+1

train_data = np.zeros([360,56*46])
test_data = np.zeros([40,56*46])

for i in range(len(train_img)):
    train_data[i] = train_img[i].reshape(1,56*46) - mean_img_re

for i in range(len(test_img)):
    test_data[i] = test_img[i].reshape(1,56*46) - mean_img_re
        
def KNNClassify(k, n):
    
    X_train = np.matmul(train_data, eigen_re[:n].T)
    X_test = np.matmul(test_data, eigen_re[:n].T)
    
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, train_label)
    test_accuracy = knn.score(X_test, test_label)
    
    accuracy = test_accuracy
    
    return accuracy


print(KNNClassify(1,140))
```
